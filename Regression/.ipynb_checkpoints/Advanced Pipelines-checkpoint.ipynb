```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import FunctionTransformer
from sklearn.linear_model import LinearRegression
```

```python
class CustomImputer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.SimpleImputer = SimpleImputer(strategy='median')
    def fit(self, X, y=None):
        self.SimpleImputer.fit(X['x2'].values.reshape(-1,1))
        return self
    def transform(self, X, y=None):
        copy_X = X.copy()
        copy_X['x2'] = self.SimpleImputer.transform(copy_X['x2'].values.reshape(-1,1))
        return copy_X
```

```python
def one_hot(df):
    df_copy = df.copy()
    one_hots = pd.get_dummies(df['x4'])
    df_copy[one_hots.columns[:-1]] = one_hots[one_hots.columns[:-1]]
    keep = ['x1', 'x2', 'x3']
    keep.extend(one_hots.columns[:-1])
    return df_copy[keep]
```

```python

class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.StandardScaler = StandardScaler()
    def fit(self, X, y=None):
        self.StandardScaler.fit(X[['x1', 'x2', 'x3']])
        return self
    def transform(self, X, y=None):
        copy_X = X.copy()
        copy_X[['x1', 'x2', 'x3']] = self.StandardScaler.transform(copy_X[['x1', 'x2', 'x3']])
        return copy_X
```

```python
def get_feats(df):
    df_copy = df.copy()
    df_copy['x2_a'] = df_copy['x2']*df_copy['a']
    df_copy['x2_b'] = df_copy['x2']*df_copy['b']
    return df_copy
```

```python
pipe = Pipeline([
    ('impute', CustomImputer()),
    ('one_hot', FunctionTransformer(one_hot, validate=False)),
    ('scale', CustomScaler()),
    ('get_feats', FunctionTransformer(get_feats, validate=False)),
    ('reg', LinearRegression(copy_X=True))
])
```

```python
pipe.fit(df[['x1', 'x2', 'x3', 'x4']], df['y'])
```
