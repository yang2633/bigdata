## Theoretical Questions

##### 1. Other KNN Measures

KNN can use other distance measures. Read through the `sklearn` documentation to see what the `p` and `metric` arguments are, <a href="https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html">https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html</a>. Then look at the wikipedia entry for Minkowski distance <a href="https://en.wikipedia.org/wiki/Minkowski_distance">https://en.wikipedia.org/wiki/Minkowski_distance</a>.

In "The distance function effect on k-nearest neighbor classification for medical datasets" Hu et. al. demonstrate that the Minkowski distance can lead to horrible performance using various medical data sets, <a href="https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4978658/">https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4978658/</a>.

See how performance is affected on the iris data set when you use different distance metrics.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
print(iris.keys())
print(iris['feature_names'])
```

```python
iris_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris_df = pd.DataFrame(iris['data'], columns=iris_features)
iris_df['class'] = iris['target']
iris_df.head()
```

```python
iris_df['class'].value_counts()
```

```python
X = iris_df[iris_features].to_numpy()
y = iris_df['class'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, stratify=y, random_state = 404)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def get_precision(model, X, y, cutoff):
    probs = model.predict_proba(X)[:, 1]
    pred = 1*(probs>=cutoff)
    tp = np.sum(pred[y==1])
    fp = np.sum(pred[y==0])
    return tp/(tp+fp)

def get_recall(model, X, y, cutoff):
    probs = model.predict_proba(X)[:, 1]
    pred = 1*(probs>cutoff)
    tp = np.sum(pred[y==1])
    fn = np.sum(y[pred==0])
    return tp/(tp+fn)
```

```python
cutoff = 0.1
ps = [1,2,5,10, 100]
precisions = np.zeros((len(ps), 1))
recalls = np.zeros((len(ps), 1))
for i in range(len(ps)):
    knn = KNeighborsClassifier(3,p=ps[i])
    knn.fit(X_train, y_train)
    precisions[i, 0] = get_precision(knn, X_test, y_test, cutoff)
    recalls[i,0] = get_recall(knn, X_test, y_test, cutoff)
    print("for Mincowski distance with p = ", ps[i])
    print("    the precision is ", precisions[i,0])
    print("    the recall is ", recalls[i, 0])
```

## Applied  Questions

##### 1. Precision-Recall Trade-Off

Return to the virginica classifier we built in `Notebook 2 - Other Classification Performance Measures`.

Play around with a $k=1,5,10,20$ nearest neighbors classifier.

For each classifier plot the precision and the recall as a function of the probability cutoff. What do you notice?

```python
iris_virginica = pd.DataFrame(iris['data'], columns = iris_features)
iris_virginica['class'] = iris['target']
iris_virginica['virginica'] = 0
iris_virginica.loc[iris_virginica['class']==2, 'virginica'] = 1
iris_virginica.head()
X = iris_virginica[iris_features].to_numpy()
y = iris_virginica['virginica'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, stratify=y, random_state=404)
```

```python
ks = [1,5,10, 20]
cutoffs = np.arange(0, 1.01, 0.01)
precs = np.zeros((len(cutoffs), 4))
recs = np.zeros((len(cutoffs), 4))
j = 0
for k in ks:
    knn = KNeighborsClassifier(k)
    knn.fit(X_train, y_train)
    for i in range(len(cutoffs)):
        precs[i, j] = get_precision(knn, X_test, y_test, cutoffs[i])
        recs[i, j] = get_recall(knn, X_test, y_test, cutoffs[i])
    j += 1
```

```python
for j in range(4):
    plt.figure(figsize=(12, 8))
    plt.plot(cutoffs, precs[:,j], label="Precision")
    plt.plot(cutoffs, recs[:,j], label="Recall")
    plt.xlabel("Cutoff")
    plt.ylabel("Performance Measure")
    plt.legend(fontsize=14)
    plt.title("Precision-Recall Trade-off for $k=$" + str(ks[j]) + "Neighbors", fontsize=18)
```
